```python
## Import necessary libraries
import pandas as pd
import numpy as np
```

```python
## Read the datasets
books=pd.read_csv("BX-Books.csv",sep=";",error_bad_lines=False,encoding='latin-1')
users=pd.read_csv("BX-Users.csv",sep=";",error_bad_lines=False,encoding='latin-1')
ratings=pd.read_csv("BX-Book-Ratings.csv",sep=";",error_bad_lines=False,encoding='latin-1')
```

```python
## Choose only required features
books = books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher','Image-URL-L']]
```

```python
books.shape
```

```python
## Rename books data features
books.rename(columns={"Book-Title":'title',
                     "Book-Author":'author',
                     'Year-Of-Publication':'year',
                     'Publisher':'publisher',
                      'Image-URL-L': 'image_url_l'},inplace=True)
```

```python
## Rename users data features
users.rename(columns={"User-ID":'user_id',
                     "Location":'location',
                     'Age':'age'
                       },inplace=True)
```

```python
## Users dataset head
users.head()
```

```python
## Features present in the books rating dataset
ratings.columns
```

```python
## Users Dataset size
ratings.shape
```

```python
## Rename book rating data features
ratings.rename(columns={"User-ID":'user_id',
                     "Book-Rating":'rating'
                       },inplace=True)
```

```python
## book rating dataset head
ratings.head()
```

```python
## All three datasets shape comparison
print(f"Books dataset: {books.shape}")
print(f"Users dataset: {users.shape}")
print(f"Book-rating dataset: {ratings.shape}")
```

```python
ratings['user_id'].value_counts()
```

```python
## Store only those users who had atleast rated more than 200 books
x = ratings['user_id'].value_counts() > 200
```

```python
x[x].shape
```

* Out of '1149780' users only '899' users rated more than 200 books 

```python
y = x[x].index
y
```

```python
## Filter records 
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
## Now join ratings dataset with books dataset
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.shape
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.head() 
```

```python
# Filter those books which got at least 50 rating of user
final_rating = final_rating[final_rating['num_of_rating'] >= 50 ]
```

```python
final_rating.head()
```

```python
## Drop duplicates
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
## Check missing values
final_rating.isnull().sum()
```

```python
## Create Pivot table
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```
